```python
# Like to view ratio
df['like_view_ratio'] = df['likes'] / df['views']

# Comment toview ratio
df['comment_view_ratio'] = df['comments'] / df['views']

# Share to view ratio
df['share_view_ratio'] = df['shares'] / df['views']

# Subscriber growth ratio
df['sub_growth_ratio'] = df['subscribersGained'] / df['views']

# Average retention 
df['avg_retention'] = df['averageViewPercentage'] / 100

df[['views','likes','comments','shares','like_view_ratio','comment_view_ratio','share_view_ratio']].head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['day'], df['views'], color='blue')
plt.xlabel('Day')
plt.ylabel('Views')
plt.title('Video Views Over Time')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['day'], df['like_view_ratio'], label='Like/View', color='green')
plt.plot(df['day'], df['comment_view_ratio'], label='Comment/View', color='orange')
plt.plot(df['day'], df['share_view_ratio'], label='Share/View', color='red')
plt.xlabel('Day')
plt.ylabel('Engagement Ratio')
plt.title('Engagement Ratios Over Time')
plt.legend()
plt.show()
```

```python
top_days = df[df['like_view_ratio'] > 0.1]  # tweak threshold
top_days[['day','views','like_view_ratio','comment_view_ratio','share_view_ratio']]
```

```python
df['day'] = pd.to_datetime(df['day'])

# first 3 days
first_day = df['day'].min()
three_days_later = first_day + pd.Timedelta(days=3)

early_df = df[df['day'] < three_days_later]
early_avg = early_df[['like_view_ratio','comment_view_ratio','share_view_ratio']].mean()
print("Average engagement ratios in first 3 days:\n", early_avg)
```

```python
engagement_cols = ['like_view_ratio','comment_view_ratio','share_view_ratio']

overall_avg = df[df['views'] > 0][engagement_cols].mean()
print("Overall average engagement ratios:\n", overall_avg)
```

```python
import matplotlib.pyplot as plt

ratios = ['like_view_ratio','comment_view_ratio','share_view_ratio']
early_values = [0.025, 0.0, 0.0]
overall_values = [0.012297, 0.000702, 0.005248]

x = range(len(ratios))
plt.bar(x, early_values, width=0.4, label='First 3 Days', align='center')
plt.bar([i + 0.4 for i in x], overall_values, width=0.4, label='Overall', align='center')
plt.xticks([i + 0.2 for i in x], ['Likes','Comments','Shares'])
plt.ylabel('Engagement Ratio')
plt.title('Early vs Overall Engagement Ratios')
plt.legend()
plt.show()
```

```python
# Convert to datetime 
df['day_dt'] = pd.to_datetime(df['day'])

# Group by day and mean of retention
daily_retention = df.groupby('day_dt')['averageViewPercentage'].mean().reset_index()

plt.figure(figsize=(12,5))
plt.plot(daily_retention['day_dt'], daily_retention['averageViewPercentage'], color='purple')
plt.xlabel('Date')
plt.ylabel('Average View %')
plt.title('Video Retention Over Time (Daily Average)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['day'], df['subscribersGained'], label='Gained', color='blue')
plt.plot(df['day'], df['subscribersLost'], label='Lost', color='red')
plt.xlabel('Day')
plt.ylabel('Subscribers')
plt.title('Subscriber Changes Over Time')
plt.legend()
plt.show()
```

```python
import numpy as np

# Replace 0 views with NaN to avoid dividing by zero
df['like_view_ratio'] = df['likes'] / df['views'].replace(0, np.nan)
df['comment_view_ratio'] = df['comments'] / df['views'].replace(0, np.nan)
df['share_view_ratio'] = df['shares'] / df['views'].replace(0, np.nan)

# Now the ratios will be NaN instead of inf
# When taking mean, NaNs are automatically ignored
overall_avg = df[['like_view_ratio','comment_view_ratio','share_view_ratio']].mean()
print(overall_avg)
```
